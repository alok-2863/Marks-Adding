# **Importing Libraries**

```python
import pandas as pd
```

# **Reading CSV File**

```python
# Read the CSV file into a DataFrame
data_frame = pd.read_csv('Mini Project - Marks Adding.csv')
print("Done")
```

# **Start Program**

```python
# Group the data by roll number and sum the marks
grouped_data = data_frame.groupby('Roll Num')['Marks'].sum().reset_index()
print("Done")

#performs a group-by operation on the DataFrame based on the 'RollNumber' 
#column, calculates the sum of the 'Marks' column for each group, and returns
#a new DataFrame (grouped_data) with the roll numbers and their corresponding
#sums of marks.'''
```

# **Saving data to new CSV file**

```python
# Save the grouped data to a new CSV file
grouped_data.to_csv('Added Marks.csv', index=False)
print("Done")
```

# **Complete Program**

```python
import pandas as pd

# Read the CSV file into a DataFrame
data_frame = pd.read_csv('Mini Project - Marks Adding.csv')

# Group the data by roll number and sum the marks
grouped_data = data_frame.groupby('Roll Num')['Marks'].sum().reset_index()

# Save the grouped data to a new CSV file
grouped_data.to_csv('Added Marks.csv', index=False)

print("Done")
```
